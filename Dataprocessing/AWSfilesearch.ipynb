```python
import boto3 
import pandas as pd
import os
from itertools import product

#load access key
HOME = os.path.expanduser('~')
KEYPATH = "SWEMLv2.0/AWSaccessKeys.csv"
ACCESS = pd.read_csv(f"{HOME}/{KEYPATH}")

BUCKET_NAME = 'national-snow-model'


S3 = boto3.resource('s3',aws_access_key_id=ACCESS['Access key ID'][0],
    aws_secret_access_key=ACCESS['Secret access key'][0],
    )

BUCKET = S3.Bucket(BUCKET_NAME)
```

## Code to look up files

```python
for object_summary in BUCKET.objects.filter(Prefix="SWEMLv2.0/data/TrainingDFs/SouthernRockies/"):
    print(object_summary.key)
```

## Code to download folders

```python
def folderload(awspaths, path):

    #Snotel metafile
    dpath = f"{HOME}/SWEMLv2.0/{path}"
 
    if not os.path.exists(dpath):
        os.makedirs(dpath, exist_ok=True)

    #get a list of all files in dpath
    files = []
    for object_summary in BUCKET.objects.filter(Prefix=awspaths):
        files.append(object_summary.key)

    for file in files:   
        print(file)
        filepath = os.path.dirname(f"{HOME}/{file}")
        if not os.path.exists(filepath):
            os.makedirs(filepath, exist_ok=True)
        key = f"{file}"            
        S3.meta.client.download_file(BUCKET_NAME, key,f"{HOME}/{file}")
```

```python
  #key files
awspath = 'SWEMLv2.0/data/TrainingDFs'
path = "data/TrainingDFs"

folderload(awspath, path)
```
