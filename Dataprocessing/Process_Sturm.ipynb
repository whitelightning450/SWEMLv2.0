### Data Combiner

```python
import os
import pandas as pd
from tqdm import tqdm

def append_column_from_parquet(file1, file2, column_name, output_file):
    # Read the first Parquet file
    df1 = pd.read_parquet(file1)
    
    # Check if the specified column exists in the first DataFrame
    if column_name not in df1.columns:
        raise ValueError(f"Column '{column_name}' not found in {file1}")
    
    # Extract the specified column and the 'cell_id' column
    extracted_column = df1[['cell_id', column_name]]
    
    # Read the second Parquet file
    df2 = pd.read_parquet(file2)
    
    # Merge the extracted column with the second DataFrame
    combined_df = pd.merge(df2, extracted_column, on="cell_id", how="left")
    
    # Save the combined DataFrame to a new Parquet file
    combined_df.to_parquet(output_file)
    
    return combined_df

def process_all_parquet_files(region_list, column_name):
    HOME = os.path.expanduser('~')
    for region in region_list:
        input_directory1 = f"{HOME}/SWEMLv2.0/data/TrainingDFs/{region}/300M_Resolution/Sturm_Seasonality_PrecipVIIRSGeoObsDFs/20_fSCA_Thresh"
        input_directory2 = f"{HOME}/SWEMLv2.0/data/TrainingDFs/{region}/300M_Resolution/Vegetation_Seasonality_PrecipVIIRSGeoObsDFs/20_fSCA_Thresh"
        output_directory = f"{HOME}/SWEMLv2.0/data/TrainingDFs/{region}/300M_Resolution/Vegetation_Sturm_Seasonality_PrecipVIIRSGeoObsDFs/20_fSCA_Thresh"
        
        # Create output directory if it doesn't exist
        os.makedirs(output_directory, exist_ok=True)
        
        # Get list of all Parquet files in the directory
        parquet_files1 = [f for f in os.listdir(input_directory1) if f.endswith('.parquet')]
        parquet_files2 = [f for f in os.listdir(input_directory2) if f.endswith('.parquet')]
        
        # Progress bar for overall processing
        for parquet_file1, parquet_file2 in tqdm(zip(parquet_files1, parquet_files2), desc=f"Processing Parquet Files for {region}"):
            input_file1 = os.path.join(input_directory1, parquet_file1)
            input_file2 = os.path.join(input_directory2, parquet_file2)
            output_file = os.path.join(output_directory, f"Vegetation_Sturm_{parquet_file1}")
            
            # Append column from first Parquet file to second
            combined_df = append_column_from_parquet(input_file1, input_file2, column_name, output_file)
            
            # Print the first few rows of the combined DataFrame
            print(combined_df.head())

# Example usage
region_list = ['Northwest', 'SouthernRockies', 'Southwest']
column_name = "sturm_value"  # Specify the column to extract from the first file

# Process all Parquet files for the given regions and column
process_all_parquet_files(region_list, column_name)
```

## Randomly Check Data

```python
import os
import pandas as pd
from tqdm import tqdm

def count_vegetation_values(input_directory):
    # Get list of all Parquet files in the directory
    parquet_files = [f for f in os.listdir(input_directory) if f.endswith('.parquet')]
    
    value_counts = {i: 0 for i in range(1, 21)}
    total_rows = 0
    
    # Progress bar for overall processing
    for parquet_file in tqdm(parquet_files, desc="Processing Parquet Files"):
        input_file = os.path.join(input_directory, parquet_file)
        
        # Load Current Data
        current_df = pd.read_parquet(input_file)
        
        # Count total rows
        total_rows += len(current_df)
        
        # Count values from 1 to 20 in 'vegetation_value' column
        if 'vegetation_value' in current_df.columns:
            for value in range(1, 21):
                value_counts[value] += (current_df['vegetation_value'] == value).sum()
        else:
            print(f"Warning: 'vegetation_value' column not found in {parquet_file}")
    
    # Output the results
    print(f"Total rows (expected values): {total_rows}")
    for value in range(1, 21):
        print(f"Total count of '{value}': {value_counts[value]}")

# Example usage
HOME = os.path.expanduser('~')
region_list = ['Northwest', 'Southwest', 'SouthernRockies']
for region in region_list:
    input_directory = f"{HOME}/SWEMLv2.0/data/TrainingDFs/{region}/300M_Resolution/Vegetation_Sturm_Seasonality_PrecipVIIRSGeoObsDFs/20_fSCA_Thresh"
    count_vegetation_values(input_directory)
```

```python
from rasterio.enums import Resampling

def resample_raster(src_path, dst_path, scale_factor):
    with rasterio.open(src_path) as src:
        # Calculate the new transform with larger grid cells
        transform = src.transform * src.transform.scale(scale_factor, scale_factor)
        
        # Update profile with the new transform, and width and height that cover the same extent
        profile = src.profile
        profile.update(transform=transform, width=int(src.width / scale_factor), height=int(src.height / scale_factor))
        
        # Read data and resample to the new shape
        data = src.read(
            out_shape=(src.count, int(src.height / scale_factor), int(src.width / scale_factor)),
            resampling=Resampling.bilinear
        )
        
        # Write the resampled data to a new file
        with rasterio.open(dst_path, 'w', **profile) as dst:
            dst.write(data)
# Example usage
HOME = os.path.expanduser('~')
src_vegetation_file = f"{HOME}/SWEMLv2.0/USA_NALCMS_landcover_2020_30m.tif"
resampled_vegetation_file = f"{HOME}/SWEMLv2.0/USA_NALCMS_landcover_2020_120m.tif"

resample_raster(src_vegetation_file, resampled_vegetation_file, scale_factor=4)  # Adjust scale_factor as needed
```

```python
import os
import pandas as pd
import numpy as np
import rasterio
from shapely.geometry import shape

# Function to sample Sturm data using centroids
def sample_sturm_data2(sturm_file, centroids):
    with rasterio.open(sturm_file) as src:
        transform = src.transform
        sturm_data = src.read(1)
        
        values = []
        nan_count = 0
        for lon, lat in centroids:
            # Transform coordinates to the raster's coordinate reference system
            row, col = ~transform * (lon, lat)
            row, col = int(row), int(col)
            # Check if the indices are within the bounds of the raster
            if 0 <= row < sturm_data.shape[0] and 0 <= col < sturm_data.shape[1]:
                value = sturm_data[row, col]
                values.append(value)
            else:
                values.append(np.nan)  # Append NaN if coordinates are out of bounds
                nan_count += 1
    return values, nan_count

# Function to append sampled values to a target Parquet file
def append_sturm_values_to_parquet(parquet_file, sturm_values, output_directory):
    try:
        df = pd.read_parquet(parquet_file)
    except Exception as e:
        print(f"Error reading Parquet file {parquet_file}: {e}")
        return

    if len(df) != len(sturm_values):
        print(f"Length of values ({len(sturm_values)}) does not match length of DataFrame index ({len(df)}) for file {parquet_file}")
        return

    df['sturm_value'] = sturm_values
    output_file_name = "Sturm_" + os.path.basename(parquet_file)
    output_file = os.path.join(output_directory, output_file_name)
    try:
        df.to_parquet(output_file, index=False)
        print(f"Updated Parquet file saved to {output_file}")
    except Exception as e:
        print(f"Error writing Parquet file {output_file}: {e}")
    return output_file
```

```python
import os
import numpy as np
import rasterio
from rasterio.mask import mask
from shapely.geometry import box
from rasterio.features import shapes
from shapely.geometry import shape

# Function to sample Sturm data using centroids
def sample_sturm_data(sturm_file, centroids):
    with rasterio.open(sturm_file) as src:
        transform = src.transform
        sturm_data = src.read(1)
        
        values = []
        nan_count = 0
        for lon, lat in centroids:
            # Transform coordinates to the raster's coordinate reference system
            row, col = ~transform * (lon, lat)
            row, col = int(row), int(col)
            # Check if the indices are within the bounds of the raster
            if 0 <= row < sturm_data.shape[0] and 0 <= col < sturm_data.shape[1]:
                value = sturm_data[row, col]
                values.append(value)
            else:
                values.append(np.nan)  # Append NaN if coordinates are out of bounds
                nan_count += 1
    return values, nan_count

# Function to clip the Sturm .tif file
def clip_sturm_tif(sturm_file, bbox, output_file):
    with rasterio.open(sturm_file) as src:
        bbox_geom = [box(*bbox)]
        out_image, out_transform = mask(src, bbox_geom, crop=True)
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_image.shape[1],
            "width": out_image.shape[2],
            "transform": out_transform
        })
        with rasterio.open(output_file, "w", **out_meta) as dest:
            dest.write(out_image)
    print(f"Clipped Sturm .tif file saved to {output_file}")

# Function to append sampled values to a target .tif file
def append_sturm_values_to_tif(target_tif, sturm_values, output_file):
    with rasterio.open(target_tif) as src:
        meta = src.meta
        data = src.read(1)
        
        # Create an output array with the same shape as the input
        output_data = np.zeros_like(data, dtype=np.float32)
        
        # Assign the sampled Sturm values to the corresponding cells
        count = 0
        for (row, col), value in zip(np.ndindex(data.shape), sturm_values):
            if not np.isnan(value):
                output_data[row, col] = value
            else:
                output_data[row, col] = data[row, col]  # Keep original value if NaN
            count += 1
        
        meta.update(dtype=rasterio.float32)
        
        with rasterio.open(output_file, 'w', **meta) as dst:
            dst.write(output_data, 1)
    print(f"Updated target .tif file saved to {output_file}")

# Directories and files
HOME = os.path.expanduser('~')
target_tif_file = f"{HOME}/SWEMLv2.0/data/ASO/Northwest/Processed_300M_SWE/ASO_300M_20160329.tif"
sturm_file = f"{HOME}/SWEMLv2.0/SnowClass_NA_300m_10_Clip.tif"
output_directory = f"{HOME}/SWEMLv2.0/Dataprocessing/SWEMLv2.0"

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Generate centroids based on the target .tif file's coordinates
with rasterio.open(target_tif_file) as src:
    transform = src.transform
    mask_data = src.dataset_mask()
    shapes_generator = shapes(mask_data, transform=transform)
    centroid_coords = []
    for geom, value in shapes_generator:
        geom_shape = shape(geom)
        centroid_coords.append((geom_shape.centroid.x, geom_shape.centroid.y))

# Sample Sturm data using the centroids
sturm_values, nan_count = sample_sturm_data(sturm_file, centroid_coords)
print(f"Total number of NaN values sampled: {nan_count}")

# Append sampled Sturm values to the target .tif file
output_target_tif = os.path.join(output_directory, "Updated_" + os.path.basename(target_tif_file))
append_sturm_values_to_tif(target_tif_file, sturm_values, output_target_tif)

# Clip the Sturm .tif file
bbox = [src.bounds.left, src.bounds.bottom, src.bounds.right, src.bounds.top]
output_sturm_tif = os.path.join(output_directory, "Clipped_" + os.path.basename(sturm_file))
clip_sturm_tif(sturm_file, bbox, output_sturm_tif)
```
