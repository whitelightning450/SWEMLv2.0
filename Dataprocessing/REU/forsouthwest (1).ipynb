```python
def calculate_wbt(temp_c, rh):
    # Empirical formula to approximate wet bulb temperature
    wbt = temp_c * np.arctan(0.151977 * np.sqrt(rh + 8.313659)) + \
          np.arctan(temp_c + rh) - np.arctan(rh - 1.676331) + \
          0.00391838 * (rh ** (3 / 2)) * np.arctan(0.023101 * rh) - 4.686035
    return wbt
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

def calculate_wbt(temp_c, rh):
    # Empirical formula to approximate wet bulb temperature
    wbt = temp_c * np.arctan(0.151977 * np.sqrt(rh + 8.313659)) + \
          np.arctan(temp_c + rh) - np.arctan(rh - 1.676331) + \
          0.00391838 * (rh ** (3 / 2)) * np.arctan(0.023101 * rh) - 4.686035
    return wbt

# Define the original bounding box coordinates
min_x = -2303499.25
max_x = 2303500.75
min_y = -1919500.375
max_y = 1919499.625

# Subset the dataset to the bounding box and select data for January 2012
bbox_ds = combined_ds.sel(x=slice(min_x, max_x), y=slice(min_y, max_y), time=slice('2012-01-01', '2012-01-31'))
print("Bounding Box and January 2012 Subset Dataset Structure:")
print(bbox_ds)

# Assuming temperature is in Kelvin and specific humidity in kg/kg
# Convert temperature to Celsius for the calculation
temp_c = bbox_ds['temperature'] - 273.15

# Check sample dry bulb temperature values (in Celsius)
sample_temp_c = temp_c.isel(time=0, x=slice(0, 10), y=slice(0, 10)).values
print("Sample Dry Bulb Temperature Values (Celsius):")
print(sample_temp_c)

# Check sample specific humidity values
sample_humidity = bbox_ds['specific_humidity'].isel(time=0, x=slice(0, 10), y=slice(0, 10)).values
print("Sample Specific Humidity Values (kg/kg):")
print(sample_humidity)

# Calculate relative humidity more accurately
def calculate_relative_humidity(temp_c, specific_humidity):
    # Constants for water vapor pressure calculation
    A = 6.1121  # hPa
    B = 17.502
    C = 240.97  # degrees Celsius
    # Calculate saturation vapor pressure in hPa
    saturation_vapor_pressure = A * np.exp((B * temp_c) / (temp_c + C))
    # Calculate actual vapor pressure in hPa
    actual_vapor_pressure = specific_humidity * 1013.25 / (0.622 + 0.378 * specific_humidity)
    # Calculate relative humidity
    rh = (actual_vapor_pressure / saturation_vapor_pressure) * 100
    return rh

sample_rh_accurate = calculate_relative_humidity(sample_temp_c, sample_humidity)
print("Sample Relative Humidity Values (Accurate, %):")
print(sample_rh_accurate)

# Recalculate wet bulb temperature using the accurate relative humidity
sample_wbt_recalculated = calculate_wbt(sample_temp_c, sample_rh_accurate)
print("Recalculated Wet Bulb Temperature Values (Celsius):")
print(sample_wbt_recalculated)

# Calculate Wet Bulb Temperature (WBT) for the entire dataset
rh = calculate_relative_humidity(temp_c, bbox_ds['specific_humidity'])
wbt = xr.apply_ufunc(calculate_wbt, temp_c, rh,
                     input_core_dims=[[], []],
                     output_core_dims=[[]],
                     vectorize=True,
                     dask='parallelized',
                     output_dtypes=[float])

# Add WBT to the dataset
bbox_ds['wet_bulb_temperature'] = wbt

# Classify precipitation type based on WBT
precip_type = xr.where(wbt <= 0, 'snow', 'rain')

# Add precipitation type to the dataset
bbox_ds['precipitation_type'] = precip_type

# Print the structure of the updated dataset
print(bbox_ds)
```
