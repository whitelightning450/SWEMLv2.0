```python
def subset_dataset(ds, year, month):
    # Define the bounding box coordinates
    min_x = -2303499.25
    max_x = 2303500.75
    min_y = -1919500.375
    max_y = 1919499.625

    # Define the time range for the month
    start_date = f'{year}-{month:02d}-01'
    if month == 12:
        end_date = f'{year+1}-01-01'
    else:
        end_date = f'{year}-{month+1:02d}-01'

    # Subset the dataset to the bounding box and the time range
    subset_ds = ds.sel(x=slice(min_x, max_x), y=slice(min_y, max_y), time=slice(start_date, end_date))

    return subset_ds
```

```python
def calculate_wbt(temp_c, rh):
    wbt = temp_c * np.arctan(0.151977 * np.sqrt(rh + 8.313659)) + \
          np.arctan(temp_c + rh) - np.arctan(rh - 1.676331) + \
          0.00391838 * (rh ** (3 / 2)) * np.arctan(0.023101 * rh) - 4.686035
    return wbt

def calculate_relative_humidity(temp_c, specific_humidity):
    A = 6.1121  # hPa
    B = 17.502
    C = 240.97  # degrees Celsius
    saturation_vapor_pressure = A * np.exp((B * temp_c) / (temp_c + C))
    actual_vapor_pressure = specific_humidity * 1013.25 / (0.622 + 0.378 * specific_humidity)
    rh = (actual_vapor_pressure / saturation_vapor_pressure) * 100
    return rh
```

```python
def add_wbt_and_precip_phase(ds):
    # Assuming temperature is in Kelvin and specific humidity in kg/kg
    temp_c = ds['temperature'] - 273.15
    specific_humidity = ds['specific_humidity']

    # Calculate relative humidity
    rh = calculate_relative_humidity(temp_c, specific_humidity)

    # Calculate Wet Bulb Temperature (WBT)
    wbt = xr.apply_ufunc(calculate_wbt, temp_c, rh,
                         input_core_dims=[[], []],
                         output_core_dims=[[]],
                         vectorize=True,
                         dask='parallelized',
                         output_dtypes=[float])

    # Add WBT to the dataset
    ds['wet_bulb_temperature'] = wbt

    # Classify precipitation type based on WBT
    precip_type = xr.where(ds['wet_bulb_temperature'] <= 0, 'snow', 'rain')

    # Add precipitation type to the dataset
    ds['precipitation_type'] = precip_type

    return ds
```
